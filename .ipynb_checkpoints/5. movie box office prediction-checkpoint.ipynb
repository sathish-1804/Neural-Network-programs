```python
import pandas as pd
import numpy as np
```

```python
import tensorflow as tf
tf.compat.v1.executing_eagerly_outside_functions()
```

```python
from sklearn.model_selection import train_test_split
from keras.layers import Input, Embedding, GRU, Dense
```

```python
from keras.models import Model
from keras.callbacks import EarlyStopping
```

```python
df = pd.read_csv("bollywood_box_clean.csv")
df
```

```python
x = df[["movie_opening","movie_weekend","movie_firstweek"]]
y = df["movie_total_worldwide"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# input_shape = (x_train.shape[1], 1)
# input_layer = Input(shape=input_shape)
# gru_layer = GRU(64, return_sequences=True, dropout=0.2, 
# recurrent_dropout=0.2)(input_layer)
# gru_layer = GRU(32, return_sequences=True, dropout=0.2, 
# recurrent_dropout=0.2)(gru_layer)
# gru_layer = GRU(16, dropout=0.2, recurrent_dropout=0.2)(gru_layer)
# output_layer = Dense(1, activation='linear')(gru_layer)
# model = Model(input_layer, output_layer)
```

```python
model = tf.keras.Sequential([
    GRU(64, return_sequences=True, dropout=0.2, recurrent_dropout=0.2, input_shape=(x_train.shape[1], 1)),
    GRU(32, return_sequences=True, dropout=0.2, recurrent_dropout=0.2),
    GRU(16, dropout=0.2, recurrent_dropout=0.2),
    Dense(1, activation='linear')
])

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5)

model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test), callbacks=[early_stopping])
```

```python
test_loss = model.evaluate(x_test, y_test) 
print(f'Test loss: {test_loss}')
```
