```python
import tensorflow as tf 
import cv2 
import tensorflow_hub as hub 
import numpy as np 
import PIL 
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image as krs_image

def load_img(path_to_img):
    img = krs_image.load_img(path_to_img, target_size=(224, 224))
    # Convert the PIL image to a NumPy array
    img_array = krs_image.img_to_array(img)
    # Add a batch dimension and scale pixel values to [0, 1]
    img_tensor = tf.expand_dims(img_array, axis=0) / 255.0
    return img_tensor
```

```python
content_path = "images1.png" 
style_path = "images2.png" 
 
dog_image = load_img(content_path) 
style_image = load_img(style_path) 
 
dog = cv2.imread(content_path) 
style = cv2.imread(style_path)
```

```python
def tensor_to_image(tensor): 
    tensor = tensor*255 
    tensor = np.array(tensor, dtype=np.uint8) 
    if np.ndim(tensor)>3: 
        assert tensor.shape[0] == 1 
        tensor = tensor[0] 
    return PIL.Image.fromarray(tensor) 
```

```python
hub_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/1') 
stylized_image = hub_model(tf.constant(dog_image), tf.constant(style_image))[0] 
op = tensor_to_image(stylized_image) 
op 
```
